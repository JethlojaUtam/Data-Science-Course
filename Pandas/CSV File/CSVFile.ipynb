### 👉GroupBy (Very Important)

#### ➡️ Single Condition

```python
df.groupby("race/ethnicity")["math score"].mean()
```

##### ✔️ Groups data and performs calculation.

#### ➡️ Multiple Condition

```python
df.groupby("race/ethnicity").agg({
    "math score": ["mean", "max"],
    "writing score": "mean"
})
```

##### ✔️ Groups data and performs calculation.
##### ✔️ agg(): Applies multiple calculations at once.

### 👉Apply, Map and Replace

#### ➡️ apply()

```python
df["reading score"].apply(lambda x : x - 5)
```

##### ✔️ Applies a function to each value.

#### ➡️ map()

```python
df["race/ethnicity"].map({"group A":1, "group B":2})
```

##### ✔️ Maps old values to new values.

#### ➡️ replace()

```python
df["lunch"].replace("standard", "free/reduced", inplace=True)
```

##### ✔️ Replaces specific values.
