# 📊 Matplotlib (Python Visualization Library)

## 📦 Import Matplotlib

```python
import matplotlib.pyplot as plt 
```

## 🧱 Matplotlib Plot Types (Types of Charts)
### 1️⃣ Line Plot
### 2️⃣ Bar Chart
### 3️⃣ Scatter Plot
### 4️⃣ Histogram
### 5️⃣ Pie Chart

## 1️⃣ Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 25]
plt.plot(x,y)
plt.show()
```

#### 🔹 plt.plot(): Draws a line plot.
#### 🔹 plt.show(): Displays the graph.

```python
x = [1, 3, 5, 7, 9, 11]
y = [10, 25, 15, 20, 35, 40]
plt.plot(x,y)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sale Trend")
plt.show()
```

#### 🔹 plt.xlabel(): Sets label for X-axis.
#### 🔹 plt.ylabel(): Sets label for Y-axis.
#### 🔹 plt.title(): Sets title of the graph.

```python
y1 = []
y2 = []
x = range(-100, 100, 10)
for i in x:
    y1.append(i**2)
for i in x:
    y2.append(-i**2)
plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.ylim(-2000, 2000)
plt.title("Multiple Line Plot")
plt.axhline(0)
plt.axvline(0)
plt.savefig("Multilineplot.jpg")
plt.show()
```

#### 🔹 plt.xlim(): Sets X-axis range.
#### 🔹 plt.ylim(): Sets Y-axis range.
#### 🔹 plt.savefig(): Saves graph as image.
#### 🔹 plt.axhline(): Draw the horizontal line.
#### 🔹 plt.axvline(): Draw the vertical line.

```python
x1 = [1, 2, 3]
y1 = [2, 4, 1]
plt.plot(x1,y1, label="line 1")

x2 = [1, 2, 3]
y2 = [4, 1, 3]
plt.plot(x2,y2, label="line 2")

plt.xlabel("x - axis")
plt.ylabel("y - axis")
plt.title("Two Line Graph")
plt.legend()
plt.show()
```

#### 🔹 plt.legend(): Shows labels for multiple lines.

## 2️⃣ Bar Chart

```python
value = [5, 6, 3, 7, 2]
name = ["A", "B", "C", "D", "E"]
plt.bar(name, value, color="green")
plt.show()
```

#### 🔹 plt.bar(): Draws a bar chart.
#### 🔹 color: Changes graph color.

```python
name = ["A", "B", "C", "D", "E", "F"]
mark = [50, 67, 30, 45, 90, 70]
plt.bar(name,mark)
plt.xlabel("Name")
plt.ylabel("Mark")
plt.title("Student Mark")
plt.show()
```

## 3️⃣ Scatter Plot

```python
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
plt.scatter(x,y)
plt.show()
```

#### 🔹 plt.scatter(): Draws a scatter plot.

```python
x = [10, 20, 30, 40, 50, 60]
y = [100, 90, 85, 30, 60, 70]
plt.scatter(x,y)
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title("Scatter Plot")
plt.savefig("ScatterPlot.png")
plt.show()
```

## 4️⃣ Histogram

```python
mark = [45, 55, 60, 70, 80, 85, 90, 95, 60, 75]
plt.hist(mark)
plt.show()
```

#### 🔹 plt.hist(): Draws histogram.

```python
mark = [95, 60, 90, 50, 88, 55, 60, 40]
plt.hist(mark, bins=5)
plt.xlabel("Marks")
plt.ylabel("Grade")
plt.title("Student Mark Graph")
plt.show()
```

## 5️⃣ Pie Chart

```python
label = ["Python", "Java", "C++"]
size = [50, 30, 20]
plt.pie(size, labels=label)
plt.show()
```

#### 🔹 plt.pie(): Draws pie chart.

```python
label = ["Python", "Java", "C++", "DotNET", "AI/ML"]
size = [25, 15, 30, 20, 10]
plt.pie(size, labels=label, autopct="%1.1f%%")
plt.title("Subject Distribution")
plt.show()
```

#### 🔹 autopct: Automatically label the wedges of a pie chart with their numerical percentage value. 
#### 🔹 labels: Add labels to the pie chart with the labels parameter.

```python
label = ["Python", "Java", "C++", "DotNET", "AI/ML"]
size = [25, 15, 30, 20, 10]
ex = (0, 0, 0, 0, 0.3)
plt.pie(size, labels=label, autopct="%1.1f%%", explode=ex, shadow=True)
plt.title("Subject Distribution")
plt.show()
```

#### 🔹 explode: The technique of separating (offsetting) one or more slices from the main body of the pie chart.
#### 🔹 shadow: This boolean attribute adds a shadow effect below the rim of the pie. 

## Customization Of Plot (Line Plot)

```python
x = [1, 2, 3, 4, 5, 6]
y = [20, 40, 10, 50, 20, 60]
plt.plot(x,y,color="green",linestyle="dashed",linewidth=3,marker="o",markerfacecolor="blue",markersize=12,label="Sales")
plt.xlim(1,8)
plt.ylim(10,80)
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Daily Sales Report")
plt.legend(loc="upper left")
plt.show()
```

#### 🔹 linestyle: Changes the line style.
#### 🔹 linewidth: Controls thickness of the line.
#### 🔹 marker: Adds points on the line.
#### 🔹 markersize: Changes size of markers.
#### 🔹 label: Gives name to a line.
#### 🔹 plt.legend(): Shows legend box.
