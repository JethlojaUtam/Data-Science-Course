# 📊 CSV File With Matplotlib

## 🔰 Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 🔰 Load CSV File

```python
df = pd.read_csv("mobile_project.csv")
```

```python
df
```

```python
df.head()
```

## 🧱 Matplotlib Plot Types (Types of Charts)
### 1️⃣ Line Plot
### 2️⃣ Bar Chart
### 3️⃣ Scatter Plot
### 4️⃣ Histogram
### 5️⃣ Pie Chart

## 1️⃣ Line Plot

```python
plt.plot(df["price_range"], df["ram"])
plt.xlabel("Mobile Price")
plt.ylabel("Ram")
plt.title("Line Plot")
plt.show()
```

## 2️⃣ Bar Chart

```python
plt.bar(df["price_range"], df["ram"])
plt.xlabel("Price")
plt.ylabel("Ram")
plt.title("Mobile Price vs Ram")
plt.show()
```

## 3️⃣ Scatter Plot

```python
plt.scatter(df["three_g"], df["px_height"])
plt.xlabel("Three G")
plt.ylabel("Height")
plt.title("Scatter Plot")
plt.show()
```

```python
plt.scatter(df["blue"], df["price_range"])
plt.xlabel("Blue")
plt.ylabel("Price")
plt.title("Scatter Plot")
plt.show()
```

## 4️⃣ Histogram

```python
plt.hist(df["ram"])
plt.xlabel("Ram")
plt.ylabel("Price")
plt.title("Distribution of Ram")
plt.show()
```

## 5️⃣ Pie Chart

```python
a = df["price_range"].value_counts()
```

```python
a
```

```python
l = ("Lower Mediam Range", "Higher Mediam Range", "Higher Range", "Lower Range")
```

```python
l
```

```python
plt.pie(a.values, labels=l)
plt.show()
```

```python
plt.pie(a.values, labels=l, autopct="%1.1f%%")
plt.title("Subject Distribution")
plt.show()
```

```python
ex = (0.3, 0, 0, 0)
```

```python
plt.pie(a.values, labels=l, autopct="%1.1f%%", explode=ex, shadow=True)
plt.title("Subject Distribution")
plt.show()
```

```python
price_count = df["price_range"].value_counts()
colors = ["lightblue", "lightgreen", "orange", "pink"]
explode = [0.1, 0.2, 0, 0]
```

```python
plt.pie(price_count, labels=price_count.index, autopct="%1.1f%%", colors=colors, explode=explode)
plt.title("Price Distribution")
plt.show()
```

## 6️⃣ Box Plot

```python
plt.boxplot(df["price_range"])
plt.ylabel("Price")
plt.title("Box Plot of Price")
plt.show()
```
