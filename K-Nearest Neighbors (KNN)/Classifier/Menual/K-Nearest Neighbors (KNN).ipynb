# 🤖 KNN (K-Nearest Neighbors) Classifier
### 👉 KNN (K-Nearest Neighbors) is a Machine Learning algorithm that makes predictions by looking at nearby data points.
### 👉 KNN predicts the result by looking at the nearest data points.
### 👉 The model checks nearest neighbors and predicts the class.
### What is KNN?
#### 👉 KNN is a machine learning algorithm that predicts output based on the nearest data points.
### What is K in KNN?
#### 👉 K represents the number of nearest neighbors used for prediction.

## 🔷 KNN Classifier
#### 👉 Output = category
#### 👉 Used when output is a class.
### Example:
#### Pass / Fail
#### Spam / Not Spam
#### Disease / No diseas

## 🔷 Concepts:

## 🔷 Distance Calculation (Important)
#### 👉 Uses KNN distance measure.
#### 👉 Measures how far two points are.
### Euclidean Distance


## 🔷 Example idea

#### | Study Hours | Pass |
#### | ----------- | ---- |
#### | 2           | Fail |
#### | 3           | Fail |
#### | 4           | Pass |
#### | 5           | Pass |
#### New student:
#### StudyHours = 4.2
- #### KNN will decide by looking at the nearest students
#### 👉 Pass or Fail

## 🔷 How does KNN work?
### 👉 The algorithm finds the nearest data points and predicts based on them.

### Step-by-step:
#### 1️⃣ Take new data point
#### 2️⃣ Find K nearest neighbors
#### 3️⃣ Calculate distance
#### 4️⃣ Take majority vote

## 🔷 KNN Important Parameter
#### 👉 Output = number

### Example:
#### K value
#### K = number of neighbors
#### KNeighborsClassifier(n_neighbors=5)

## 🔷 Advantages
#### ✔️ Simple algorithm
#### ✔️ Easy to understand
#### ✔️ Works well with small data

## 🔷 Disadvantages
#### ❌ Slow with large data
#### ❌ Sensitive to noise
#### ❌ Needs feature scaling

## 🔰 Import Library

```python
from sklearn.neighbors import KNeighborsClassifier
```

## 🔰 Data

```python
input_data = [[2], [3], [4], [5], [6]]
output_data = [0,0,1,1,1]
```

## 🔰 Model

```python
model = KNeighborsClassifier(n_neighbors = 5) 
```

```python
model.fit(input_data,output_data)
```

## 🔰 Prediction

```python
pred = model.predict([[4.5]])
print(pred)
```
