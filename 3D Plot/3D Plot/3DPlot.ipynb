# 📊 3D Plot

## 🔰 Import Library

```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as axes3D
```

## 🔍 Types of 3D Plots
### 1. 3D Scatter Plot
### 2. 3D Line Plot
### 3. 3D Surface Plot
### 4. 3D Bar Plot¶

## 1. 3D Scatter Plot

```python
math = [78, 88, 65, 92, 55]
science = [85, 90, 70, 95, 60]
english = [72, 91, 60, 94, 58]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(math, science, english)
ax.set_xlabel("Math")
ax.set_ylabel("Science")
ax.set_zlabel("English")
plt.title("3D Scatter Plot")
plt.show()
```

## 2. 3D Line Plot

```python
days = [1, 2, 3, 4, 5]
sales = [100, 200, 350, 400, 250]
profite = [20, 30, 60, 45, 80]
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111, projection="3d")
ax1.plot(days, sales, profite)
ax1.set_xlabel("Days")
ax1.set_ylabel("Sales")
ax1.set_zlabel("Profite")
plt.title("3D Line Plot")
plt.show()
```

## 3. 3D Surface Plot

```python
import numpy as np
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x,y = np.meshgrid(x,y)
z = x**2 + y**2
fig = plt.figure()
ax2 = fig.add_subplot(111, projection="3d")
ax2.plot_surface(x, y, z)
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_zlabel("Z")
plt.title("3D Surface Plot")
plt.show()
```

## 4. 3D Bar Plot

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = [0, 0, 0]
```

```python
da = db = [0.5, 0.5, 0.5]
dc = [100, 150, 200]
```

```python
fig = plt.figure()
ax3 = fig.add_subplot(111, projection="3d")
ax3.bar3d(a, b, c, da, db, dc)
ax3.set_xlabel("Product")
ax3.set_ylabel("Month")
ax3.set_zlabel("Sales")
plt.title("3D Bar Plot")
plt.show()
```
