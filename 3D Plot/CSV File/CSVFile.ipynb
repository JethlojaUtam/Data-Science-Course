# 📊 3D Plot With CSV File

## 🔰 Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as axes3D
```

## 🔰 CSV File Load

```python
df = pd.read_csv("mobile_project.csv")
```

```python
df
```

```python
df.head()
```

## 🔍 Types of 3D Plots
### 1. 3D Scatter Plot
### 2. 3D Line Plot
### 3. 3D Surface Plot
### 4. 3D Bar Plot

## 1. 3D Scatter Plot

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1, projection="3d")
ax.scatter(df["fc"], df["ram"], df["price_range"])
ax.set_xlabel("Front Camera")
ax.set_ylabel("Ram")
ax.set_zlabel("Price Range")
ax.set_title("3D Scatter Plot")
plt.show()
```

## 2. 3D Line Plot

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection="3d")
ax.plot(df["fc"], df["price_range"], df["ram"])
ax.set_xlabel("Front camera")
ax.set_ylabel("Price Range")
ax.set_zlabel("Ram")
ax.set_title("3D Line Plot")
plt.show()
```

## 3. 3D Surface Plot

```python
import numpy as np
```

```python
x = np.array(df["fc"])
y = np.array(df["price_range"])
z = np.array(df["ram"])
x, y = np.meshgrid(x, y)
z = x + y
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection="3d")
ax.plot_surface(x, y, z)
ax.set_xlabel("Front Camera")
ax.set_ylabel("Price Range")
ax.set_zlabel("Ram")
ax.set_title("3D Surface Plot")
plt.show()
```

### 4. 3D Bar Plot

```python
x = df["fc"]
y = df["price_range"]
z = np.zeros(len(df))
dx = np.ones(len(df)) * 0.5
dy = np.ones(len(df)) * 0.5
dz = df["ram"]
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection="3d")
ax.bar3d(x, y, z, dx, dy, dz)
ax.set_xlabel("Front Camera")
ax.set_ylabel("Price Range")
ax.set_zlabel("Ram")
ax.set_title("3D Bar Plot")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.bar3d(x, y, z, dx, dy, dz, color="skyblue")
ax.set_xlabel("Front Camera")
ax.set_ylabel("Price Range")
ax.set_zlabel("Ram")
ax.set_title("3D Bar Plot")
plt.show()
```
