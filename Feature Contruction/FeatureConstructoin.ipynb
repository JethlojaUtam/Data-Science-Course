# 🧠 Feature Construction

### 👉 The process of creating new meaningful features (columns) from raw data so that the Machine Learning model learns better.

## 🔍 Example (Simple Understanding)
#### Raw data:
##### Age = 25
##### Height = 170 cm
##### Weight = 70 kg

#### Constructed feature:
##### BMI = Weight / (Height/100)^2

## 🔰 Required the columns

```python
df1 = pd.read_csv("mobile_project.csv")[["sc_h", "sc_w"]]
```

```python
df1
```

## 🔰 Drop the null value

```python
df1.dropna()
```

## 🔰 Feature Construction

```python
df1["sc_area"] = df1["sc_h"] * df1["sc_w"]
```

```python
df1
```

```python
df1.min()
```

```python
df1.max()
```

```python
def mymobile(screen):
    if screen >= 228:
        return "Heavy"
    elif screen >= 114 :
        return "Mediam"
    else:
        return "Light"
```

```python
df1["sc_type"] = df1["sc_area"].apply(mymobile)
```
