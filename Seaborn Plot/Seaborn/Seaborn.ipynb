# 📊 Seaborn (Python Visualization Library)

#### 👉 Seaborn is used to create beautiful and easy statistical graphs from data.

#### 👉 Seaborn is a high-level visualization library used to create attractive statistical graphs easily from data.

## 📦 Seaborn Import

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 🧱 Seaborn Plot Types (Types of Graphs)
### 🔹 Relational Plots
### 🔹 Categorical Plots
### 🔹 Distribution Plots
### 🔹 Regression Plots
### 🔹 Matrix Plots

## 🔰 DataFrame Sample Data

```python
df = pd.DataFrame({
    "age": [18, 20, 22, 24, 26, 28, 30, 32],
    "mark": [55, 60, 65, 70, 75, 80, 85, 90],
    "gender": ["M", "F", "F", "M", "M", "F", "M", "F"]
})
```

```python
df
```

### 1️⃣ Distribution Plot (dist)

```python
sns.distplot(df["mark"], bins=5)
plt.title("Distribution of Mark")
plt.show()
```

```python
sns.distplot(df["mark"])
plt.title("Distribution of Mark")
plt.show()
```

#### 🔹 sns.histplot(): Shows how data is spread.
#### 🔹 bins: number of bars

### 2️⃣ Dis Plot (dis)

```python
sns.displot(df["mark"])
plt.title("Dis Plot")
plt.show()
```

### 3️⃣ Histogram Plot (hist)

```python
sns.histplot(df["mark"])
plt.title("histogram of Mark")
plt.show()
```

#### 🔹 sns.histplot(): Shows how data is spread

### 4️⃣ Bar Plot (bar)

```python
sns.barplot(x = "gender", y = "mark", data=df)
plt.title("Average mark of gender")
plt.show()
```

```python
sns.barplot(x = "gender", y = "mark", data=df, hue="age")
plt.title("Average mark of gender")
plt.show()
```

#### 🔹 barplot(): Shows average value per category.

### 5️⃣ Scatter Plot (scatter)

```python
sns.scatterplot(x = "gender", y = "mark", data=df)
plt.title("Average mark of gender")
plt.show()
```

```python
sns.scatterplot(x = "gender", y = "mark", data=df, hue="age")
plt.title("Average mark of gender")
plt.show()
```

#### 🔹 sns.scatterplot(): Shows relationship between two variables using dots.

### 6️⃣ Box Plot (box)

```python
sns.boxplot(x = "gender", y = "mark", data=df)
plt.title("box plot of mark")
plt.show()
```

```python
sns.boxplot(x = "age", y = "mark", data=df)
plt.title("box plot of mark")
plt.show()
```

#### 🔹 Shows median, spread, outliers.

### 7️⃣ Heatmap (heat)

```python
cor = df[["age", "mark"]].corr()
```

```python
cor
```

```python
sns.heatmap(cor)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.heatmap(cor, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

#### 🔹 heatmap(): Color-based matrix.
#### 🔹 annot: True shows values.

### 8️⃣ Violin Plot (violin)

```python
sns.violinplot(x= "mark", y = "age", data=df)
plt.title("Violin Plot of Marks")
plt.show()
```

```python
sns.violinplot(y = "age", data=df)
plt.title("Violin Plot of Marks")
plt.show()
```

```python
sns.violinplot(x= "mark", data=df)
plt.title("Violin Plot of Marks")
plt.show()
```

#### 🔹 violinplot(): Distribution + shape

### 9️⃣ Categorical Plot (cat)

```python
sns.catplot(df)
plt.show()
```

```python
g = sns.catplot(df)
g
g.set_xticklabels(rotation=90)
```

### 🔟 Joint Plot (joint)

```python
sns.jointplot(df)
plt.title("Joint Plot")
plt.show()
```

```python
sns.jointplot(data=df, x="age", y="mark")
plt.title("Joint Plot")
plt.show()
```

###  1️⃣1️⃣ Line Plot (line)

```python
sns.lineplot(x="mark", y="age", data=df)
plt.title("Marks Trend by Age")
plt.show()
```

### 1️⃣2️⃣ Matrix Plots (Correlation / Table Data)

```python
import numpy as np
data = np.array([[1,2],[3,4]])
sns.heatmap(data, annot=True)
plt.show()
```

## 🎨 Style, Theme, Palette (IMPORTANT)

```python
sns.barplot(x = "gender", y = "mark", data=df)
plt.title("Average mark of gender")
plt.show()
sns.set_style("darkgrid")
sns.set_theme(font_scale=1.2)
sns.set_palette("pastel")
```

```python
sns.distplot(df["mark"])
plt.title("Distribution of Mark")
plt.show()
sns.set_style("dark")
sns.set_theme(font_scale=1.2)
sns.set_palette("pastel")
```

```python
sns.boxplot(x = "gender", y = "mark", data=df)
plt.title("box plot of mark")
plt.show()
sns.set_style("white")
sns.set_theme(font_scale=1)
sns.set_palette("pastel")
```

#### 🔹 sns.set_style(): Changes overall look of plots.
#### 🔹 "white"
#### 🔹 "whitegrid"
#### 🔹 "dark"
#### 🔹 "darkgrid"

#### 🔹 sns.set_palette(): Changes color palette, Changes graph colors.
#### 🔹 deep
#### 🔹 muted
#### 🔹 pastel
#### 🔹 Set1, Set2

#### 🔹 Theme(): Controls overall look and text size.
