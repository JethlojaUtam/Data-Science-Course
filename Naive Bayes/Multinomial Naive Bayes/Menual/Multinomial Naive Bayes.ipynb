# 🧠 Naive Bayes Algorithm
### 👉 Naive Bayes probability-based algorithm is based on Bayes' Theorem.
### 👉 Basic idea:
### 👉 “Given input features, what is the probability of the class?”

## 🔷 What is “Naive”?

## “Naive” is:

#### 👉 The algorithm assumes that the features are independent

### Example:

- #### Email with “free” and “offer” words → independent arguments

#### 👉 (Real life is not true, but the algorithm is fast and effective)

## 🔷 Naive Bayes Concept

### Explanation:

- #### P(C|X) = Posterior → final probability (answer)
- #### P(X|C) = Likelihood
- #### P(C) = Prior probability
- #### P(X) = Evidence

## 🔶 Multinomial Naive Bayes
#### 👉 Best for text data.
#### 👉 Text data / word count companion uses.
#### 👉 Mostly NLP I use.
### Example:
#### Email spam detection
#### Word count based features

### 📌 Example:

#### Email spam check:
#### | Message        | Spam?    |
#### | -------------- | -------- |
#### | free offer now | Spam     |
#### | win money now  | Spam     |
#### | hello friend   | Not Spam |
#### | meeting today  | Not Spam |

### 👉 New message: "free money"
### → Output: Spam

## 🔰 Import Library

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

## 🔰 Data

```python
texts = ["free offer now", "win money now", "hello friend", "meeting today", "earn money fast"]
labels = [1, 1, 0, 0, 1]
```

```python
vectorizer = CountVectorizer()
```

```python
x = vectorizer.fit_transform(texts)
```

## 🔰 Model

```python
model = MultinomialNB()
```

```python
model.fit(x,labels)
```

## 🔰 Prediction

```python
test = vectorizer.transform(["free money"])
```

```python
pred = model.predict(test)
```
