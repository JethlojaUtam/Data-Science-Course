# 🧱 Feature Scaling with Real CSV (Normalization (Min–Max Scaling))

## 🔰 Import Libary

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

## 🔰 Load CSV File

```python
df = pd.read_csv("mobile_project.csv")
```

```python
df
```

```python
df.head()
```

```python
df.info()
```

## 🔰 Use Of Scaler Variable

```python
battery = df[["battery_power"]]
```

```python
battery
```

```python
battery.min()
```

```python
battery.max()
```

## 🔰 Object and Transform

```python
scaler = MinMaxScaler()
```

```python
battery_scaler = scaler.fit_transform(battery)
```

## 🔰 Convert To Array

```python
battery_scaler = np.array(battery_scaler)
```

```python
battery_scaler
```
