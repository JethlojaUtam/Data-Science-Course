# 📐 Feature Scaling In Normalization (Min–Max Scaling)

### 👉 Normalization scales data between 0 and 1.

### 👉 Converts values into range 0 to 1.

### 👉 Normalization scales data between 0 and 1 using min-max formula.

## 📐 Normalization Formula:

## 🔢 Normalization Example:-

##### Marks = [50, 60, 70, 80, 90]
##### Min = 50
##### Max = 90

##### For 70:
##### (70 - 50) / (90 - 50) = 20 / 40 = 0.5

#### 👉 New value = 0.5

## 📐 Normalization Range:
#### 👉 Mostly 0 to 1
#### 👉 Sometimes -1 to 1

## 📐 Use Of Normalization: 
### ✔️ KNN
### ✔️ K-Means
### ✔️ Neural Networks
### ✔️ Image processing

## 📐 Example Of Normalization :-

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler_data = scaler.fit_transform([[50], [60], [65], [70], [75]])
```
