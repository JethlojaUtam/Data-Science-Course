# 🧱 Feature Scaling with Real CSV (Standardization (Z-score Scaling))

## 🔰 Import Libary

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

## 🔰 Load CSV File

```python
df = pd.read_csv("mobile_project.csv")
```

## 🔰 Use Of Scaler Variable

```python
battery = df[["battery_power"]]
```

```python
battery
```

```python
battery.min()
```

```python
battery.max().values[0]
```

```python
battery.min().values[0]
```

## 🔰 Object and Transform

```python
scaler = StandardScaler()
```

```python
battery_scaler = scaler.fit_transform(battery)
```

## 🔰 Convert To Array

```python
battery_scaler = np.array(battery_scaler)
```

```python
battery_scaler
```

## 🔰 Chaeck The Value

```python
np.min(battery_scaler)
```

```python
np.max(battery_scaler)
```
