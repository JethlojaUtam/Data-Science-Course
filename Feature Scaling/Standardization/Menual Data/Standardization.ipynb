# 📐 Feature Scaling In Standardization (Z-score Scaling)

### 👉 Standardization scales data around mean 0 with unit variance.

### 👉 Converts data so that mean = 0 and standard deviation = 1.

### 👉 Standardization scales data to mean 0 and standard deviation 1.

## 📐 Standardization Formula:

## 🔢 Standardization Example:-

##### Marks = [50, 60, 70, 80, 90]
##### Mean = 70
##### Std = 14.14 (approx)

##### For 80:
##### (80 - 70) / 14.14 ≈ 0.7

#### 👉 New value = 0.7

## 📐 Standardization Range:
#### 👉 No fixed range
#### 👉 Values can be negative or positive

## 📐 Use Of Standardization
### ✔️ Linear Regression
### ✔️ Logistic Regression
### ✔️ SVM
### ✔️ PCA

## 📐 Example Of Standardization :-

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```
