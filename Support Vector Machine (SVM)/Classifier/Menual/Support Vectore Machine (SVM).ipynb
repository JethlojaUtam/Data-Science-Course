# 🧠 SVM (Support Vector Machine) Classifier
### 👉 SVM (Support Vector Machine) is a Machine Learning algorithm that finds the best boundary to divide data into two classes.
### 👉 SVM finds the best line (or boundary) that separates data into different classes.
### 👉 SVM is a machine learning algorithm that finds the optimal boundary to separate data into classes.

#### 👉 Data points closest to the separating boundary.

## 🔷 Concepts:

## 🔷 Key idea
#### 👉 SVM tries to create the widest possible margin between classes.
### SVM finds:
- #### Hyperplane → separating line
- #### Margin → distance between classes
- #### Support vectors → closest points to boundary

## 🔷 How does SVM work?
#### 👉 SVM separates classes using the best possible boundary.
### Steps:
#### 1️⃣ Plots data points
#### 2️⃣ Finds the best separating line between classes
#### 3️⃣ Creates maximum margin
#### 4️⃣ Makes predictions

## 🔷 SVM Classification
#### 👉 Output = category
### Examples:
- #### Spam / Not Spam
- #### Pass / Fail
- #### Disease / No disease
#### 👉 Used when output is a class.

## 🔷 Kernel Types (Very Important)
#### 👉 SVM kernels make decision boundary changes.

### Kernel --> Meaning
#### Linear --> straight line
#### Polynomial --> curved line
#### RBF (Radial) --> complex boundary
#### Sigmoid --> neural-like

#### 👉 Kernel decides the shape of decision boundary.

## 🔷 Important Parameters
### Kernel:
#### svm.SVC(kernel='rbf')
#### svm.SVC(kernel='linear')
#### svm.SVC(kernel='polynomial')
#### svm.SVC(kernel='sigmoid')
### C (Regularization):
#### svm.SVC(C=1.0)
#### 👉 C controls overfitting.

## 🔷 Advantages
#### ✔️ Works well in high-dimensional data
#### ✔️ Good for classification
#### ✔️ Strong theoretical foundation

## 🔷 Disadvantages
#### ❌ Slow for large datasets
#### ❌ Hard to tune parameters
#### ❌ Needs feature scaling

## 🔰 Import Library

```python
from sklearn import svm
```

## 🔰 Data

```python
input_data = [[2], [3], [4], [5], [6]]
output_data = [0,0,1,1,1]
```

## 🔰 Model

```python
model = svm.SVC(kernel = "linear")
```

```python
model.fit(input_data,output_data)
```

## 🔰 Prediction

```python
pred = model.predict([[4.5]])
print(pred)
```
