# 📈 Logistic Regression

### 👉 Logistic Regression is a classification algorithm.
### 👉 It gives output like Yes / No, 0 / 1, Pass / Fail.

### 👉 Logistic Regression predicts a category (class).

## 🔍 Use Of Logistic Regression? 

#### - Pass / Fail
#### - Spam / Not Spam
#### - Disease / No Disease
#### - Loan Approved / Rejected

## 🧠 Core Idea (Simple)
### Logistic Regression uses the Sigmoid function:

- Output → 0 to 1
- This value = probability

## 🧠 When to Use Which? (Golden Rule)
- #### Output is category (Yes/No) → Logistic Regression

## 🧩 Import Library

```python
from sklearn.linear_model import LogisticRegression
```

## 🧩 Sample Data

```python
input_data = [[1], [2], [3], [4], [5]]
output_data = [0, 0, 0, 1, 1]
```

## 🧩 Model

```python
model = LogisticRegression()
```

```python
model.fit(input_data, output_data)
```

```python
prediction = model.predict([[3]])
```

```python
probability = model.predict_proba([[3]])
```
