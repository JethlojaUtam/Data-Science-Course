# 📈 Polynomial Regression

### 👉 Polynomial regression is used when data follows a curved relationship instead of a straight line.
### 👉 Polynomial features convert straight data into curve form for better prediction.
### 👉 If graph is curved → use polynomial regression.

## 🔷 Advantages
#### ✔️ Handles curved data
#### ✔️ More accurate than linear
#### ✔️ Flexible model

## 🔷 Disadvantages
#### ❌ Overfitting risk
#### ❌ High degree = complex
#### ❌ Slow for big data

## 🔷 When to Use Polynomial Regression?
#### ✔️ Data curved
#### ✔️ Linear regression fit is not
#### ✔️ Non-linear relation

## 🔷 Degree Concept (Important)
### Degree -> Meaning
##### 1 -> Linear
##### 2 -> Curve (square)
##### 3 -> More curve
##### High degree -> Complex curve

## When to use?
### When data shows a curved relationship.

## 🔷 Import Library

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## 🔷 Data

```python
input_data = np.array([[1],[2],[3],[4],[5]])
output_data = np.array([35,45,60,80,110])
```

## 🔷 Polynomial convert

```python
poly = PolynomialFeatures(degree=2)
input_poly = poly.fit_transform(input_data)
```

## 🔷 Model

```python
model = LinearRegression()
model.fit(input_poly, output_data)
```

## 🔷 Prediction

```python
pred = model.predict(poly.transform([[6]]))
print(pred)
```
