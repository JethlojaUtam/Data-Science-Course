# 🔷 Label Encoding

### 👉 Give each category a unique number.

### 👉 Converts categories into numbers.

## ❗ Label Encoding ONLY use

#### ✔ Gender, Yes/No, Pass/Fail
#### ✔️ Target variable (Yes/No)
#### ✔️ Binary categorical data

## 🧾 Import Library

```python
from sklearn.preprocessing import LabelEncoder
```

## 🧾 Data Load

```python
label = LabelEncoder()
```

```python
data = ["Male", "Female", "Female", "Male", "Female", "Male"]
```

```python
data
```

```python
label_encoder = label.fit_transform(data)
```
