# 🔴 Ordinal Encoding

### 👉 Education, Rating, Size (order matters)

### 👉 Ordinal Encoding converts text to numbers while keeping the correct order.

### 👉 Used when categories have a meaningful order.

## ✅ Import Library

```python
import pandas as pd
```

## ✅ DataFrame

```python
data = { "Performance": ["High", "Medium", "Low", "High", "Medium"] }
```

```python
data
```

## ✅ Method : 1

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
order = { "Low":0 ,"Medium":1 ,"High":2}
```

```python
order
```

```python
df["Performance_Encoder"] = df["Performance"].map(order)
```

```python
df["Performance_Encoder"]
```

```python
df
```

## ✅ Method : 2

## ✅ Import Library

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
```

```python
encoders = OrdinalEncoder(categories=[["Low", "Medium", "High"]])
```

```python
df["Performance_Encoder_2"] = encoders.fit_transform(df[["Performance"]])
```
