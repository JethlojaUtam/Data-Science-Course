# 🔤 Ordinal Encoding with Real CSV

## 🔰 Import Library

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

## 🔰 Load CSV File

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.head()
```

## 🔰 Convert To OrdinalEncoding

```python
df["Embarked"].unique()
```

```python
df["Embarked"].fillna("S", inplace=True)
```

```python
coder = OrdinalEncoder(categories=[["S", "C", "Q"]])
```

```python
df["Embarked_encoder"] = coder.fit_transform(df[["Embarked"]])
```
