# 🔷 One-Hot Encoding

### 👉 Creates separate columns for each category.

### 👉 Converts categories into numbers while keeping order.

### 👉 Each category → separate column

## 🧠 When to use One-Hot?
#### ✔ City
#### ✔ Country
#### ✔ Product type
#### ✔ Department

## 🔰 Import Library

```python
import pandas as pd
```

## 🔰 Convert To OneHotEncoding

```python
data = { "brand": ["Samsung", "Apple", "MI", "Samsung"] }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

## 🔰 Method : 1

```python
encoder = pd.get_dummies(df["brand"])
```

```python
encoder
```

```python
data_encoder = pd.concat([df, encoder], axis=1)
```

```python
data_encoder
```

## 🔰 Method : 2

## 🔰 Import Library

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
encoders = OneHotEncoder(sparse_output=False)
```

```python
encoders_data = encoders.fit_transform(df[["brand"]])
```

```python
encoders_data
```

```python
data_encoders = pd.DataFrame(encoders_data, columns=encoders.get_feature_names_out(["brand"]))
```

```python
data_encoders
```
