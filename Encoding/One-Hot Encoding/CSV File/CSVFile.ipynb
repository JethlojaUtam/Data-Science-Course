# 🔤 One Hot Encoding with Real CSV

## 🔰 Import Library

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

## 🔰 Load CSV File

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.head()
```

## 🔰 Convert To OneHotEncoding

## 🔰 Method : 1 (Simple Method)

```python
embarked_encoder = pd.get_dummies(df["Embarked"], prefix="Embarked")
```

```python
embarked_encoder
```

```python
df = pd.concat([df, embarked_encoder], axis=1)
```

```python
df
```

## 🔰 Method : 2 (Complex Method)

```python
encoder = OneHotEncoder()
```

```python
encoders = OneHotEncoder(sparse_output=False)
```

```python
onehotcoder = encoders.fit_transform([["Sex", "Embarked"]])
```

```python
df1 = pd.DataFrame(onehotcoder, columns=encoders.get_feature_names_out(["Sex", "Embarked"]))
```

```python
df1
```

```python
df = pd.concat([df.reset_index(drop=True).drop(columns=["Sex", "Embarked"]), df1.reset_index(drop=True)], axis=1)
```
