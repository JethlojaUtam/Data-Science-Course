# 🚆 Train & Test & Split

## 🧱 What is Train Data?

#### 👉 Train Data is the part of the dataset that is used to train the model (learn patterns).

#### 👉 Train data teaches the model.

### Example: By looking at student marks, the model learns how marks are calculated.

## 🧪 What is Test Data?

#### 👉 Test data is the part of the dataset that is there to check if the model has learned well.

#### 👉 Test data checks the model.

### Important: The model was not seen during test data training.

## ❓ Why is Split necessary?

#### 👉 If we give the entire data to training then:
- Model remembers answers
- Gives poor performance on real world data
- 👉 This is called Overfitting

#### 👉 Without split, model only memorizes instead of learning.

### 👉 Rule of Thumb:
- Most common = 80% Train, 20% Test
